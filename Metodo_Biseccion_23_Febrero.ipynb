```python
import math
import matplotlib.pyplot as plt 
from tabulate import tabulate 
%matplotlib inline

xVerd=14.7802
m, g, v, t, xl, xu, xr, fA, fB=68.1, 9.8, 40.0, 10.0, 12.0, 16.0, 0.0, 1.0, 1.0
valores, funcion, Error, tabla, raices, tablaB=[],[],[],[],[],[]

print "Dando valores a c, podemos hallar el resultado de f(c)\n"
for i in range(1,24):
    coef=i*1
    fc=((g*m)/coef)*(1-math.pow(math.e,-(coef/68.1)*10.0))-v
    if(coef>0 and coef%4==0):
        valores.append(coef)
        funcion.append(fc)
        tabla.append([coef, fc])

print tabulate(tabla, headers=["Valores de c", "f(c)"], tablefmt="fancy_grid")

plt.title("COEFICIENTE")
plt.axis([0, 24, -10, 40])
plt.ylabel('f(c)')
plt.xlabel('c')
plt.axhline(0, color="black")
plt.plot(valores, funcion, 'm-', marker='o')
plt.plot(xVerd,0, 'cs', label='Raiz Exacta=14.7802')
plt.legend(loc = 'upper right', numpoints = 1)
plt.show()

def metodo1(num):
    fc=((g*m)/num)*(1-math.exp(-(num/68.1)*10.0))-v
    return fc
    
while ((fA*fB)!=0):
    xr=(xl+xu)/2
    fA=metodo1(xl)
    fB=metodo1(xr)
    erpv=math.fabs(((xVerd-xr)/xVerd)*100)
    tablaB.append([xr, str(erpv)+" %"])
    if ((fA*fB)<0):
        xu=xr
    if((fA*fB)>0):
         xl=xr   

print "\nUsando el metodo de biseccion, y entre mas iteraciones del mismo se hagan, se estimara una raiz mas cercana a la raiz verdadera del problema. Dicho calculo se detendra cuando el producto de las funciones f(xl)f(xr)=0\n"
print tabulate(tablaB, headers=["Raices", "Error"], tablefmt="fancy_grid")
```
