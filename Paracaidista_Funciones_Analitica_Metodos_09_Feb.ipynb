Calcular la velocidad en el tiempo t de un paracaidista en caída libre, de manera analítica y por métodos numéricos. La línea punteada representa la primera forma y la línea continua representa la segunda. 

```python
import matplotlib.pyplot as plt
%matplotlib inline

grav, mass, coef, exp = 9.8, 68.1 , 12.5 , 2.7182
time=[]
speed=[]
value= 0.0
t1 = 0 
t2 = 2
v= 0.0
vel = 0.0
listTiempo=[]
listVel=[]


for i in range (0,100):
    time.append(i)
    value = ((grav*mass)/coef)*(1-exp**(-(coef/mass)*time[i]))
    speed.append(value)

print "Problema por forma analitica"
print "  Time (s) "
print "-----------"

for x in time:
    print x

print "  Speed (m/s)  "
print "---------------"

for z in speed:
    print z

print "Problema por metodos numericos"
print "Time(s) - Speed(m/s)"
for i in range (50):
    v = vel +(grav-(coef/mass)*vel)*(t2-t1)
    print "%d - %f" %(t2,v)
    vel=v
    listVel.append(v)
    t1=t2
    listTiempo.append(t2)
    t2=t1+2
    
plt.axis([0,100,0,100])
plt.ylabel('Speed (m/s)')
plt.xlabel('Time (s)')
plt.plot(listTiempo, listVel)

plt.plot([time[0:]], [speed[0:]], 'yo')
plt.show()
```
