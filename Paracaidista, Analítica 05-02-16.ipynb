Primer laboratorio de métodos numéricos. De manera analítica, determine la velocidad de un paracaidista en caída libre a medida que pasa el tiempo.

```python
import matplotlib.pyplot as plt
%matplotlib inline

grav = 9.8
mass = 68.1
coef = 12.5
exp = 2.7182
time=[]
speed=[]
value= 0.0


for i in range (0,100):
    time.append(i)
    value = ((grav*mass)/coef)*(1-exp**(-(coef/mass)*time[i]))
    speed.append(value)
    
print "  Time (s) "
print "-----------"

for x in time:
    print x

print "  Speed (m/s)  "
print "---------------"

for z in speed:
    print z
    
plt.plot([time[0:]], [speed[0:]], 'o')
plt.ylabel('Speed (m/s)')
plt.xlabel('Time (s)')
plt.axis([0, 100, 0, 100])
plt.show()
```
